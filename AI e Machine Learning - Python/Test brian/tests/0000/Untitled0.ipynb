```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)


plt.plot(x,y1,label="sin",color="blue")

plt.xlabel("questa è x")
plt.ylabel("questa è y")
plt.show()
```
