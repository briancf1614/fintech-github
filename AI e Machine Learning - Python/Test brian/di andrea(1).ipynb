```python
from google.colab import drive
import os
path = "/content/drive/My Drive/Blockchain/file_exercise/sha256/"

def set_list_file():
  os.chdir(path)
  return os.listdir()

def check_input(file, files):
  if file in files:
    return False
  else:
    return True

def hash_file(file):
  filename= path + file
  with open(filename,"rb") as f:
    bytes = f.read()
    return sha256(bytes).hexdigest()
    # readable_hash = sha256(bytes).hexdigest();
    # return sha256(readable_hash.encode('utf-8')).hexdigest()

def hash_files(files):
  list_hash = []
  for file in files:
    list_hash.append(hash_file(file))
  return list_hash

def Merkle_tree_root(files):
  list_hash= hash_files(files)
  while((len(list_hash)%2)==0):
    # visual
    display(list_hash)
    print("\n")
    # calculator
    list_hash= calculation_root(list_hash)
  if len(list_hash) == 1:
    return list_hash
  else:
    return False

def calculation_root(list):
  new_list = []
  time = len(list)/2
  k=0
  for x in range(int(time)):
    conc_hash = list[k]+ list[k+1]
    new_list.append(sha256(conc_hash.encode('utf-8')).hexdigest())
    k = x + 2
  return new_list

def main():
   
    list_files = set_list_file()
    print(str(len(list_files))+" File: ")
    display(list_files)

    print("\nHash: ")
    root = Merkle_tree_root(list_files)

    print("Tree root: ")
    display(root)

if __name__ == "__main__":
    main()
```
