```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets 
import seaborn as sns
```

```python
housing = datasets.fetch_california_housing(as_frame=True)
df = housing.frame

display(df)
```

```python

plt.figure(figsize=(13, 10))
corr_matrix = df.corr()
display(corr_matrix)

sns.heatmap(corr_matrix, annot=True)
plt.show()
```
